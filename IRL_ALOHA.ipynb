```python
# @title
import numpy as np


def sample_box_pose(seed=None):
    x_range = [0.0, 0.2]
    y_range = [0.4, 0.6]
    z_range = [0.05, 0.05]

    rng = np.random.RandomState(seed)

    ranges = np.vstack([x_range, y_range, z_range])
    cube_position = rng.uniform(ranges[:, 0], ranges[:, 1])

    cube_quat = np.array([1, 0, 0, 0])
    return np.concatenate([cube_position, cube_quat])


def sample_insertion_pose(seed=None):
    # Peg
    x_range = [0.1, 0.2]
    y_range = [0.4, 0.6]
    z_range = [0.05, 0.05]

    rng = np.random.RandomState(seed)

    ranges = np.vstack([x_range, y_range, z_range])
    peg_position = rng.uniform(ranges[:, 0], ranges[:, 1])

    peg_quat = np.array([1, 0, 0, 0])
    peg_pose = np.concatenate([peg_position, peg_quat])

    # Socket
    x_range = [-0.2, -0.1]
    y_range = [0.4, 0.6]
    z_range = [0.05, 0.05]

    ranges = np.vstack([x_range, y_range, z_range])
    socket_position = rng.uniform(ranges[:, 0], ranges[:, 1])

    socket_quat = np.array([1, 0, 0, 0])
    socket_pose = np.concatenate([socket_position, socket_quat])

    return peg_pose, socket_pose


def sample_pick_and_place_pose(seed=None):
    """Sample initial pose for cube in pick-and-place task"""
    rng = np.random.RandomState(seed)

    # Sample cube position on table
    x_range = [-0.1, 0.3]
    y_range = [0.25, 0.75]
    z_range = [0.05, 0.5]  # Fixed height on table

    ranges = np.vstack([x_range, y_range, z_range])
    position = rng.uniform(ranges[:, 0], ranges[:, 1])

    # Fixed upright orientation (quaternion)
    orientation = np.array([1, 0, 0, 0])

    return np.concatenate([position, orientation])


def sample_platform_pose(seed=None):
    """Sample a random position for the target platform."""
    rng = np.random.RandomState(seed)
    # Adjust these ranges based on your workspace constraints
    x = rng.uniform(-0.1, 0.3)
    y = rng.uniform(0.25, 0.75)
    z = rng.uniform(0.05, 0.5)
    return np.array([x, y, z])
```

```python
# @title
from pathlib import Path
import numpy as np
import copy
### Simulation envs fixed constants
DT = 0.02  # 0.02 ms -> 1/0.2 = 50 hz
FPS = 50


JOINTS = [
    # absolute joint position
    "left_arm_waist",
    "left_arm_shoulder",
    "left_arm_elbow",
    "left_arm_forearm_roll",
    "left_arm_wrist_angle",
    "left_arm_wrist_rotate",
    # normalized gripper position 0: close, 1: open
    "left_arm_gripper",
    # absolute joint position
    "right_arm_waist",
    "right_arm_shoulder",
    "right_arm_elbow",
    "right_arm_forearm_roll",
    "right_arm_wrist_angle",
    "right_arm_wrist_rotate",
    # normalized gripper position 0: close, 1: open
    "right_arm_gripper",
]

ACTIONS = [
    # position and quaternion for end effector
    "left_arm_waist",
    "left_arm_shoulder",
    "left_arm_elbow",
    "left_arm_forearm_roll",
    "left_arm_wrist_angle",
    "left_arm_wrist_rotate",
    # normalized gripper position (0: close, 1: open)
    "left_arm_gripper",
    "right_arm_waist",
    "right_arm_shoulder",
    "right_arm_elbow",
    "right_arm_forearm_roll",
    "right_arm_wrist_angle",
    "right_arm_wrist_rotate",
    # normalized gripper position (0: close, 1: open)
    "right_arm_gripper",
]


START_ARM_POSE = [
    0,
    -0.96,
    1.16,
    0,
    -0.3,
    0,
    0.02239,
    -0.02239,
    0,
    -0.96,
    1.16,
    0,
    -0.3,
    0,
    0.02239,
    -0.02239,
]

ASSETS_DIR = '/content/gym-aloha/gym_aloha/assets' # note: absolute path

# Left finger position limits (qpos[7]), right_finger = -1 * left_finger
MASTER_GRIPPER_POSITION_OPEN = 0.02417
MASTER_GRIPPER_POSITION_CLOSE = 0.01244
PUPPET_GRIPPER_POSITION_OPEN = 0.05800
PUPPET_GRIPPER_POSITION_CLOSE = 0.01844

# Gripper joint limits (qpos[6])
MASTER_GRIPPER_JOINT_OPEN = 0.3083
MASTER_GRIPPER_JOINT_CLOSE = -0.6842
PUPPET_GRIPPER_JOINT_OPEN = 1.4910
PUPPET_GRIPPER_JOINT_CLOSE = -0.6213

MASTER_GRIPPER_JOINT_MID = (MASTER_GRIPPER_JOINT_OPEN + MASTER_GRIPPER_JOINT_CLOSE) / 2

DEFAULT_PLATFORM_POSE = [0.3, 0.4, 0.05]
PLATFORM_POSE = copy.copy(DEFAULT_PLATFORM_POSE)  # Add this with BOX_POSE

############################ Helper functions ############################


def normalize_master_gripper_position(x):
    return (x - MASTER_GRIPPER_POSITION_CLOSE) / (
        MASTER_GRIPPER_POSITION_OPEN - MASTER_GRIPPER_POSITION_CLOSE
    )


def normalize_puppet_gripper_position(x):
    return (x - PUPPET_GRIPPER_POSITION_CLOSE) / (
        PUPPET_GRIPPER_POSITION_OPEN - PUPPET_GRIPPER_POSITION_CLOSE
    )


def unnormalize_master_gripper_position(x):
    return x * (MASTER_GRIPPER_POSITION_OPEN - MASTER_GRIPPER_POSITION_CLOSE) + MASTER_GRIPPER_POSITION_CLOSE


def unnormalize_puppet_gripper_position(x):
    return x * (PUPPET_GRIPPER_POSITION_OPEN - PUPPET_GRIPPER_POSITION_CLOSE) + PUPPET_GRIPPER_POSITION_CLOSE


def convert_position_from_master_to_puppet(x):
    return unnormalize_puppet_gripper_position(normalize_master_gripper_position(x))


def normalizer_master_gripper_joint(x):
    return (x - MASTER_GRIPPER_JOINT_CLOSE) / (MASTER_GRIPPER_JOINT_OPEN - MASTER_GRIPPER_JOINT_CLOSE)


def normalize_puppet_gripper_joint(x):
    return (x - PUPPET_GRIPPER_JOINT_CLOSE) / (PUPPET_GRIPPER_JOINT_OPEN - PUPPET_GRIPPER_JOINT_CLOSE)


def unnormalize_master_gripper_joint(x):
    return x * (MASTER_GRIPPER_JOINT_OPEN - MASTER_GRIPPER_JOINT_CLOSE) + MASTER_GRIPPER_JOINT_CLOSE


def unnormalize_puppet_gripper_joint(x):
    return x * (PUPPET_GRIPPER_JOINT_OPEN - PUPPET_GRIPPER_JOINT_CLOSE) + PUPPET_GRIPPER_JOINT_CLOSE


def convert_join_from_master_to_puppet(x):
    return unnormalize_puppet_gripper_joint(normalizer_master_gripper_joint(x))


def normalize_master_gripper_velocity(x):
    return x / (MASTER_GRIPPER_POSITION_OPEN - MASTER_GRIPPER_POSITION_CLOSE)


def normalize_puppet_gripper_velocity(x):
    return x / (PUPPET_GRIPPER_POSITION_OPEN - PUPPET_GRIPPER_POSITION_CLOSE)


def convert_master_from_position_to_joint(x):
    return (
        normalize_master_gripper_position(x) * (MASTER_GRIPPER_JOINT_OPEN - MASTER_GRIPPER_JOINT_CLOSE)
        + MASTER_GRIPPER_JOINT_CLOSE
    )


def convert_master_from_joint_to_position(x):
    return unnormalize_master_gripper_position(
        (x - MASTER_GRIPPER_JOINT_CLOSE) / (MASTER_GRIPPER_JOINT_OPEN - MASTER_GRIPPER_JOINT_CLOSE)
    )


def convert_puppet_from_position_to_join(x):
    return (
        normalize_puppet_gripper_position(x) * (PUPPET_GRIPPER_JOINT_OPEN - PUPPET_GRIPPER_JOINT_CLOSE)
        + PUPPET_GRIPPER_JOINT_CLOSE
    )


def convert_puppet_from_joint_to_position(x):
    return unnormalize_puppet_gripper_position(
        (x - PUPPET_GRIPPER_JOINT_CLOSE) / (PUPPET_GRIPPER_JOINT_OPEN - PUPPET_GRIPPER_JOINT_CLOSE)
    )

SINGLE_ARM_JOINTS = [
    # absolute joint position
    "left_arm_waist",
    "left_arm_shoulder",
    "left_arm_elbow",
    "left_arm_forearm_roll",
    "left_arm_wrist_angle",
    "left_arm_wrist_rotate",
    # normalized gripper position 0: close, 1: open
    "left_arm_gripper",
]

SINGLE_ARM_ACTIONS = SINGLE_ARM_JOINTS  # Same as joints for position control

SINGLE_ARM_START_POSE = START_ARM_POSE[:8]  # Just the left arm portion
```

```python
# @title
import gymnasium as gym
import numpy as np
from dm_control import mujoco
from dm_control.rl import control
from gymnasium import spaces
# from gym_aloha.constants import DEFAULT_PLATFORM_POSE

# from gym_aloha.constants import (
#     ACTIONS,
#     ASSETS_DIR,
#     DT,
#     JOINTS,
#     SINGLE_ARM_JOINTS,
#     SINGLE_ARM_ACTIONS,
# )
from gym_aloha.tasks.sim import BOX_POSE, InsertionTask, TransferCubeTask
from gym_aloha.tasks.sim_end_effector import (
    InsertionEndEffectorTask,
    TransferCubeEndEffectorTask,
)
# from gym_aloha.utils import sample_box_pose, sample_insertion_pose, sample_pick_and_place_pose, sample_platform_pose


class AlohaEnv(gym.Env):
    # TODO(aliberts): add "human" render_mode
    metadata = {"render_modes": ["rgb_array"], "render_fps": 50}

    def __init__(
        self,
        task,
        obs_type="pixels",
        render_mode="rgb_array",
        observation_width=640,
        observation_height=480,
        visualization_width=640,
        visualization_height=480,
    ):
        super().__init__()
        self.task = task
        self.obs_type = obs_type
        self.render_mode = render_mode
        self.observation_width = observation_width
        self.observation_height = observation_height
        self.visualization_width = visualization_width
        self.visualization_height = visualization_height

        self._max_episode_steps = 500 # Set this value appropriately

        self._env = self._make_env_task(self.task)

        if self.obs_type == "state":
            raise NotImplementedError()
            self.observation_space = spaces.Box(
                low=np.array([0] * len(JOINTS)),  # ???
                high=np.array([255] * len(JOINTS)),  # ???
                dtype=np.float64,
            )
        elif self.obs_type == "pixels":
            self.observation_space = spaces.Dict(
                {
                    "top": spaces.Box(
                        low=0,
                        high=255,
                        shape=(self.observation_height, self.observation_width, 3),
                        dtype=np.uint8,
                    )
                }
            )
        elif self.obs_type == "pixels_agent_pos":
            joint_count = len(SINGLE_ARM_JOINTS) if task == "pick_and_place" else len(JOINTS)
            self.observation_space = spaces.Dict(
                {
                    "pixels": spaces.Dict(
                        {
                            "top": spaces.Box(
                                low=0,
                                high=255,
                                shape=(self.observation_height, self.observation_width, 3),
                                dtype=np.uint8,
                            )
                        }
                    ),
                    "agent_pos": spaces.Box(
                        low=-1.0,
                        high=1.0,
                        shape=(joint_count,),
                        dtype=np.float64,
                    ),
                }
            )
        elif self.obs_type == "pixels_agent_pos_obj":
            joint_count = len(SINGLE_ARM_JOINTS) if task == "pick_and_place" else len(JOINTS)
            self.observation_space = spaces.Dict(
                {
                    "pixels": spaces.Dict(
                        {
                            "top": spaces.Box(
                                low=0,
                                high=255,
                                shape=(self.observation_height, self.observation_width, 3),
                                dtype=np.uint8,
                            )
                        }
                    ),
                    "agent_pos": spaces.Box(
                        low=-1.0,
                        high=1.0,
                        shape=(joint_count,),
                        dtype=np.float64,
                    ),
                    "env_state": spaces.Box(
                        low=-1.0,
                        high=1.0,
                        shape=(6,),
                        dtype=np.float64,
                    ),
                }
            )

        # Choose appropriate action space based on task
        if task == "pick_and_place":
            self.action_space = spaces.Box(
                low=-1,
                high=1,
                shape=(len(SINGLE_ARM_ACTIONS),),  # Use single arm actions
                dtype=np.float32
            )
        else:
            self.action_space = spaces.Box(
                low=-1,
                high=1,
                shape=(len(ACTIONS),),  # Use bimanual actions
                dtype=np.float32
            )

    def render(self):
        return self._render(visualize=True)

    def _render(self, visualize=False):
        assert self.render_mode == "rgb_array"
        width, height = (
            (self.visualization_width, self.visualization_height)
            if visualize
            else (self.observation_width, self.observation_height)
        )
        if mode in ["visualize", "human"]:
            height, width = self.visualize_height, self.visualize_width
        elif mode == "rgb_array":
            height, width = self.observation_height, self.observation_width
        else:
            raise ValueError(mode)
        # TODO(rcadene): render and visualizer several cameras (e.g. angle, front_close)
        image = self._env.physics.render(height=height, width=width, camera_id="top")
        return image

    def _make_env_task(self, task_name):
        # time limit is controlled by StepCounter in env factory
        time_limit = float("inf")

        if task_name == "transfer_cube":
            xml_path = "/content/gym-aloha/gym_aloha/assets/bimanual_viperx_transfer_cube.xml"
            physics = mujoco.Physics.from_xml_path(str(xml_path))
            task = TransferCubeTask()
        elif task_name == "insertion":
            xml_path = "/content/gym-aloha/gym_aloha/assets/bimanual_viperx_insertion.xml"
            physics = mujoco.Physics.from_xml_path(str(xml_path))
            task = InsertionTask()
        elif task_name == "end_effector_transfer_cube":
            raise NotImplementedError()
            xml_path = "/content/gym-aloha/gym_aloha/assets/bimanual_viperx_end_effector_transfer_cube.xml"
            physics = mujoco.Physics.from_xml_path(str(xml_path))
            task = TransferCubeEndEffectorTask()
        elif task_name == "end_effector_insertion":
            raise NotImplementedError()
            xml_path = "/content/gym-aloha/gym_aloha/assets/bimanual_viperx_end_effector_insertion.xml"
            physics = mujoco.Physics.from_xml_path(str(xml_path))
            task = InsertionEndEffectorTask()
        else:
            raise NotImplementedError(task_name)

        env = control.Environment(
            physics, task, time_limit, control_timestep=DT, n_sub_steps=None, flat_observation=False
        )
        return env

    def _format_raw_obs(self, raw_obs):
        if self.obs_type == "state":
            raise NotImplementedError()
        elif self.obs_type == "pixels":
            obs = {"top": raw_obs["images"]["top"].copy()}
        elif self.obs_type == "pixels_agent_pos":
            obs = {
                "pixels": {"top": raw_obs["images"]["top"].copy()},
                "agent_pos": raw_obs["qpos"],
                "env_state": raw_obs["env_state"],
            }
        elif self.obs_type == "pixels_agent_pos_obj":
            obs = {
                "pixels": {"top": raw_obs["images"]["top"].copy()},
                "agent_pos": raw_obs["qpos"],
                "env_state": raw_obs["env_state"],
            }
        else:
            raise ValueError(f'Invalid obs_type {self.obs_type}')
        return obs

    def reset(self, seed=None, options=None):
        super().reset(seed=seed)

        if seed is not None:
            self._env.task.random.seed(seed)
            self._env.task._random = np.random.RandomState(seed)

        if self.task == "transfer_cube":
            BOX_POSE[0] = sample_box_pose(seed)
        elif self.task == "insertion":
            BOX_POSE[0] = np.concatenate(sample_insertion_pose(seed))
        elif self.task == "pick_and_place":
            print(f'--------- Seed: {seed}')
            BOX_POSE[0] = sample_pick_and_place_pose(seed)
            PLATFORM_POSE[0] = sample_platform_pose(seed if seed is None else seed+1)
        else:
            raise ValueError(self.task)

        raw_obs = self._env.reset()

        observation = self._format_raw_obs(raw_obs.observation)

        info = {"is_success": False}
        return observation, info

    def step(self, action):
        assert action.ndim == 1
        # TODO(rcadene): add info["is_success"] and info["success"] ?
        _, reward, _, raw_obs = self._env.step(action)

        # TODO(rcadene): add an enum
        terminated = is_success = reward == 4

        info = {"is_success": is_success}

        observation = self._format_raw_obs(raw_obs)

        truncated = False
        return observation, reward, terminated, truncated, info

    def close(self):
        pass

    def set_state(self, state):
        physics = self.unwrapped._env._physics
        with physics.reset_context():
            physics.data.qpos[:] = state['qpos']
            physics.data.qvel[:] = state['qvel']

            # Get cube position and dimensions from physics model
            cube_pos = physics.data.qpos[-7:-4]
            cube_geom_id = physics.model.name2id("red_cube", "geom")
            holder_geom_id = physics.model.name2id("holder_platform", "geom")

            cube_size = physics.model.geom_size[cube_geom_id][0]
            holder_thickness = physics.model.geom_size[holder_geom_id][2]

            # Place holder under cube
            holder_id = physics.model.name2id("holder", "body")
            physics.model.body_pos[holder_id] = [
                cube_pos[0],
                cube_pos[1],
                cube_pos[2] - (cube_size + holder_thickness)
            ]

            # Set platform position
            target_id = physics.model.name2id("target", "body")
            if 'platform_transform' in state:
                physics.model.body_pos[target_id] = state['platform_transform']['pos']
                physics.model.body_quat[target_id] = state['platform_transform']['quat']
            elif 'platform_pos' in state:
                physics.model.body_pos[target_id] = state['platform_pos']
            else:
                physics.model.body_pos[target_id] = DEFAULT_PLATFORM_POSE

        self.unwrapped.current_action = np.array(state['current_action'])

    def get_observations(self):
        import pdb; pdb.set_trace()
        return 14, {}
```

```python
# Import required libraries
import numpy as np
import torch
import gymnasium as gym
import pickle
from typing import Dict, Optional
from typing import Any
from gymnasium.wrappers import TimeLimit
from imitation.data import rollout
from stable_baselines3 import PPO
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.policies import MultiInputActorCriticPolicy, ActorCriticPolicy
from stable_baselines3.ppo import MlpPolicy
from imitation.algorithms import bc
from imitation.data import rollout
from imitation.data.wrappers import RolloutInfoWrapper
from imitation.policies.serialize import load_policy
from stable_baselines3.common.vec_env import DummyVecEnv, VecNormalize, VecEnvWrapper
from imitation.util.util import make_vec_env

SEED = 42

# Define Environment Setup
def make_env(gym_kwargs):
    """Create an instance of the environment with specific parameters."""
    env = AlohaEnv(**gym_kwargs) #AlohaEnv(**gym_kwargs)
    return env


class MyVecEnv(gym.Env):
    def __init__(self, num_envs: int, device: torch.device, gym_kwargs):
        self.num_envs = num_envs

        self.device = device
        self.state = None
        self.envs = [make_env(gym_kwargs) for _ in range(num_envs)]

        self.num_obs = self.envs[0].observation_space['env_state'].shape[0]
        self.num_actions = self.envs[0].action_space.shape[0]
        self.max_episode_length = self.envs[0]._max_episode_steps
        self.observation_space = self.envs[0].observation_space['agent_pos']
        self.action_space = self.envs[0].action_space

        self.obs_buf = torch.zeros((num_envs, self.num_obs), device=device)
        self.rew_buf = torch.zeros((num_envs,), device=device)
        self.reset_buf = torch.zeros((num_envs,), dtype=torch.bool, device=device)
        self.episode_length_buf = torch.zeros((num_envs,), dtype=torch.int32, device=device)
        self.extras = {'observations': {}}

    def reset(self, seed: int = None, options: Optional[Dict[str, Any]] = None):
        super().reset(seed=seed, options=options)
        self.state = self.observation_space.sample()
        return self.state, {}

    def step(self, action):
        reward = -np.abs(self.state - action).mean()
        self.state = self.observation_space.sample()
        return self.state, reward, False, False, {}

# Define RL Training Parameters (from config.yaml)
params = {
    "value_loss_coef": 1.0,
    "clip_param": 0.2,
    "use_clipped_value_loss": True,
    "desired_kl": 0.01,
    "entropy_coef": 0.01,
    "gamma": 0.99,
    "lam": 0.95,
    "max_grad_norm": 1.0,
    "learning_rate": 1e-5,
    "num_learning_epochs": 5,
    "num_mini_batches": 4,
    "num_steps_per_env": 24,
    "max_iterations": 1500,
    "save_interval": 50,
    "seed": 1
}

# Device setup (GPU/CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Initialize Environment obs_type="pixels_agent_pos"
gym_kwargs = {"obs_type": "pixels_agent_pos_obj", "task": "insertion", "render_mode": "rgb_array"} # "task": "insertion",
num_envs = 4  # Adjust based on available resources
# env = MyVecEnv(num_envs=num_envs, device=device, gym_kwargs=gym_kwargs)

import gymnasium as gym
from gymnasium.envs.registration import register

# Register the environment with a factory function
register(
    id="MyVecEnv-v0",
    entry_point=MyVecEnv,  # Use the factory function
    max_episode_steps=500,
)

# Now use:
env = gym.make("MyVecEnv-v0", num_envs=num_envs, device=device, gym_kwargs=gym_kwargs)
# Option A: use the `make_vec_env` helper function - make sure to pass `post_wrappers=[lambda env, _: RolloutInfoWrapper(env)]`
def _make_env():
    """Helper function to create a single environment. Put any logic here, but make sure to return a RolloutInfoWrapper."""
    _env = MyVecEnv(num_envs=num_envs, device=device, gym_kwargs=gym_kwargs)
    _env = TimeLimit(_env, max_episode_steps=500)
    _env = RolloutInfoWrapper(_env)
    return _env

venv = DummyVecEnv([_make_env for _ in range(4)])

# Initialize PPO Policy
policy_kwargs = dict(
    activation_fn=torch.nn.ELU,
    net_arch=[128, 128, 128]
)

# Train PPO Agent
def train_ppo():
    """Train PPO Agent on the custom environment."""
    model = PPO(
        policy=MlpPolicy, ## MultiInputActorCriticPolicy, ActorCriticPolicy
        env=env,
        seed=params["seed"],
        learning_rate=params["learning_rate"],
        batch_size=(params["num_steps_per_env"] * num_envs) // params["num_mini_batches"],
        ent_coef=params["entropy_coef"],
        gamma=params["gamma"],
        gae_lambda=params["lam"],
        clip_range=params["clip_param"],
        vf_coef=params["value_loss_coef"],
        max_grad_norm=params["max_grad_norm"],
        n_epochs=params["num_learning_epochs"],
        verbose=1,
        device=device,
        policy_kwargs=policy_kwargs
    )
    reward, _ = evaluate_policy(model, env, 10)
    print(f"Reward before training: {reward}")

    print("Training PPO Agent...")
    model.learn(total_timesteps=params["num_steps_per_env"] * num_envs * params["max_iterations"], progress_bar=False, log_interval=10)

    reward, _ = evaluate_policy(model, model.get_env(), 10)
    print(f"Expert reward: {reward}")

    return model

# Load Pretrained Expert
def download_expert():
    """Load a pretrained expert."""
    print("Downloading a pretrained expert.")
    expert = load_policy(
        "ppo-huggingface",
        organization="HumanCompatibleAI",
        env_name="MyVecEnv-v0",
        venv=env
    )
    return expert

# Sample Expert Transitions for IRL
def sample_expert_transitions():
    """Train an expert or load a pretrained one, then sample transitions."""
    expert = train_ppo()  # Uncomment to train your own expert
    # expert = download_expert()  # Uncomment to load a pretrained expert

    print("Sampling expert transitions...")

    rng = np.random.default_rng()
    rollouts = rollout.rollout(
        expert,
        venv,
        rollout.make_sample_until(min_timesteps=None, min_episodes=50),
        rng=rng,
    )

    # Save expert demonstrations
    with open("expert_demos.pkl", "wb") as f:
        pickle.dump(rollouts, f)

    print("Expert demonstrations saved.")
    return expert, rollout.flatten_trajectories(rollouts)

# Train Expert and Sample Transitions
expert, transitions = sample_expert_transitions()
```
